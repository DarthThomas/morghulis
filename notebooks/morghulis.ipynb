```python
%matplotlib inline

from morghulis.widerface import Wider
import numpy as np
import cv2
import imutils

from matplotlib import pyplot as plt
```

```python
widerface = Wider('/datasets/WIDER/')
```

```python
train_set = [i for i in widerface.train_set()]
val_set = [i for i in widerface.val_set()]
```

```python
crowded = [i for i in train_set if len(i.faces) > 1000]
```

```python
image = crowded[3].draw_faces(color=(0, 255, 0))
plt.figure(figsize=(15,15))
plt.imshow(image)
plt.show()
```

```python
sample = val_set[1000]
image = sample.draw_faces(color=(0, 255, 0))
plt.figure(figsize=(15,15))
plt.imshow(image)
plt.show()
```
